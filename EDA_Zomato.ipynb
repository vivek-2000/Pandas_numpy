 Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato.csv", encoding="latin-1")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
## IN Data Analysis what all things we do
# 1. Missing values
# 2. Explore about the numerical variables
# 3. Explore about categorical variables
# 4. Find the realtionship between them(features)
```

```python
# how much null value all column have
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on="Country Code", how='left')
final_df.head()
```

```python
## to check dattypes
final_df.dtypes
```

```python
country_names_array = final_df.Country.unique()
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
plt.pie(final_df.Country.value_counts().values, labels=country_names, autopct="%1.2f%%")
plt.show()
```

```python
plt.pie(final_df.Country.value_counts().values, labels=country_names_array, autopct="%1.2f%%")
plt.show()
```

```python
##pie chart for top 3 country
plt.pie(final_df.Country.value_counts().values[:3], labels=country_names_array[:3], autopct="%1.2f%%")
plt.show()
```

```python
## Observation : Zomato max records or transcation are from india after that usa & then uk
```

```python
final_df.columns
```

```python
final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size()
```

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

```python
## Observation
# 1. when rating is between 4.5 to 4.9 ---> excellent
# 2. when rating is between 4.0 to 4.4 ---> very good
# 3. when rating is between 3.0 to 3.9 ---> good
# 4. when rating is between 2.0 to 2.9 ---> average
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count",hue="Rating color", data=ratings, palette=['grey', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
## count plot # fequency of data not count
sns.countplot(x="Rating color", data=ratings)
```
